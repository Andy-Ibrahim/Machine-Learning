```python
# Import then necessary math libraries.
import numpy as np
import matplotlib.pyplot as plt
```

**Linear Functions**:
We will be using the term linear equation to to mean a weighted sum of inputs plus an offset. If theres just one input x, then this is a straight line:

\begin{equation}y=\beta+\omega x,\end{equation}

Where $\beta$ is the y-intercept of the linear equation and $\omega$ is the slope of the line. When there are 2 inputs, x1 & x2, then this equation becomes.

\begin{equation}y=\beta+\omega_1 x_1 + \omega_2 x_2.\end{equation}






```python
# define a linear function with just one input.

def linear_function_1D(x,beta,omega):
  y = beta + omega * x
  return y
```

```python
# plot the 1D linear function
# define an array of x values from 0 - 10 with increments of 0.01.
# https://numpy.org/doc/stable/reference/generated/numpy.arange.html
x = np.arange(0.0,10.0,0.01)
beta = 10.0; omega = -2.0

y = linear_function_1D(x,beta,omega)

# plot the function
fig, ax = plt.subplots()
ax.plot(x,y,'r-')
ax.set_ylim([0,10]);ax.set_xlim([0,10])
ax.set_xlabel('x'); ax.set_ylabel('y')
plt.show()
```

Lets investigage a 2D linear function


```python
def draw_2D_function(x1_mesh, x2_mesh,y):
  fig, ax = plt.subplots()
  fig.set_size_inches(7,7)
  pos = ax.contourf(x1_mesh, x2_mesh, y, levels=256 ,cmap = 'hot', vmin=-10,vmax=10.0)
  fig.colorbar(pos, ax=ax)
  ax.set_xlabel('x1');ax.set_ylabel('y1')
  levels = np.arange(-10,10,1.0)
  ax.contour(x1_mesh, x2_mesh, y, levels, cmap='winter')
  plt.show()
```

```python
def linear_function_2d(x1,x2,beta,omega1,omega2):
  y = beta + ((omega1 * x1) + (omega2 * x2))

  return y
```

```python
# Plot the 2D function

# Make 2D array of x and y points
x1 = np.arange(0.0,10.0,1)
y1 = np.arange(0.0,10.0,1)
x1,y1 = np.meshgrid(x1,x2)

beta = 0.0; omega1 = 1.0; omega2 = 0.0
y  = linear_function_2d(x1, y1, beta, omega1, omega2)

draw_2D_function(x1,y1,y)
```

```python
def linear_function_3d(x1,x2,x3,omega1,omega2,omega3,beta):
  y = beta + (x1*omega1) +  (x2*omega2) +  (x3*omega3)

  return y

```

lets compute 2 linear equations, using both individual equations and vector/matrix form and check they give the same result.

```python
beta1 = 0.5; beta2 = 0.2;

omega11 = -1.0; omega12 = 0.4; omega13 = -0.3;
omega21 = 0.1; omega22 = 0.1; omega23 = 1.2;

#define the inputs
x1 = 4; x2 = -1; x3 = 2;

y1 = linear_function_3d(x1,x2,x3,omega11,omega12,omega13, beta1)
y2 = linear_function_3d(x1,x2,x3,omega21,omega22,omega23, beta2)
print(y1, y2)
```

```python
beta_vector = np.array([[beta1],[beta2]])
omega_vector = np.array([[omega11,omega12,omega13],[omega21,omega22,omega23]])
input_vector = np.array([[4], [-1], [2]])

y_vec = beta_vector + np.matmul(omega_vector,input_vector)
print("Matrix/vector form")
print('y1= %3.3f\ny2 = %3.3f'%((y_vec[0][0],y_vec[1][0])))
```

```python
# define parameters
beta_a=3; beta_b=4; beta_c=6

omega11 = 0.5; omega12 = -0.3; omega13= 1.2
omega21 = -0.2; omega22 - 0.3; omega23 = -2.1
omega31 = 1.2; omega32 = 0.8; omega33 = 0.3

# inputs
x1_a = 2; x2_a=4; x3_a= 0

# compute the linear function for y1, y2, y3.
y1 = linear_function_3d(x1,x2,x3,omega11,omega12,omega13,beta_a)
y2 = linear_function_3d(x1,x2,x3,omega21,omega22,omega23,beta_b)
y3 = linear_function_3d(x1,x2,x3,omega31,omega32,omega33,beta_c)

print("Individual equations")
print('y1 = %3.3f\ny2 = %3.3f\ny3 = %3.3f'%((y1,y2,y3)))

```
